##### CT scans Quantification

This notebook should be in the same environement as the sample folders containing the ct images

```python
import skimage.io as io
import os
import numpy as np
import shutil
import cv2



def quantificationCT(imCT):

	ligne,colonne=imCT.shape

	imLabel=imCT*0

	for l in range(ligne):

		for c in range(colonne):

			valeurConvertir = imCT[l,c] * (3000+1050) - 1050

			if -1050<=valeurConvertir<=-950:

				imLabel[l,c]=0

			elif -950<valeurConvertir<=-120:

				imLabel[l,c]=1

			elif -120<valeurConvertir<=19:

				imLabel[l,c]=2

			elif -19<valeurConvertir<=-80:

				imLabel[l,c]=3

			elif -80<valeurConvertir<=120:

				imLabel[l,c]=4

			elif 120<valeurConvertir<=1640:

				imLabel[l,c]=5

			elif 1640<valeurConvertir<=2300:

				imLabel[l,c]=6

			elif 2300<valeurConvertir<=3000:

				imLabel[l,c]=7

	return imLabel



# Get the list of directories in the current folder
directories = [d for d in os.listdir() if os.path.isdir(d) and d.startswith('sample_')]
directories = sorted(directories)

for directory in directories:
    # Extract the number from the folder name (assuming the format 'sample_xxxx')
    folder_number = directory.split('_')[1]

    # Assuming ct.mhd is in each folder
    ct_file_path = os.path.join(directory, 'ct.mhd')

    # Load the image using SimpleITK
    img = io.imread(ct_file_path, plugin='simpleitk')

    # Annotate the image using the provided annotation function
    annotated_image = quantificationCT(img)

    # Save the result with the folder number in the filename
    annotated_ct_file_path = os.path.join(directory, 'annotated_ct')
    np.save(annotated_ct_file_path, annotated_image)
```
